**Using AND**
- Both conditions must be True
- `age = 25`
- `if age >= 18 and age < 65:`
- `    print("Working age")`


**Using OR**
- At least one condition must be True
- `day = "Saturday"`
- `if day == "Saturday" or day == "Sunday":`
- `    print("It's weekend!")`


**Using NOT**
- Reverses the condition
- `is_raining = False`
- `if not is_raining:`
- `    print("Let's go outside!")`


**Combining Multiple Conditions**
- `if age >= 18 and has_license and not is_tired:`
- `    print("You can drive")`

<div style="padding: 40px; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; font-size: 1.1em; line-height: 1.8; color: #333;">

**Basic While Loop**
- Repeats as long as condition is True
- `count = 0`
- `while count < 5:`
- `    print(count)`
- `    count = count + 1`


**Counting Down**
- `number = 5`
- `while number > 0:`
- `    print(number)`
- `    number -= 1`


**User Input Loop**
- `password = ""`
- `while password != "secret":`
- `    password = input("Enter password: ")`


**Be Careful!**
- Infinite loop: `while True:` with no break
- Always update the variable in the loop
- Use `break` to exit when needed

Working with Files
</div>
<div style="padding: 40px; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; font-size: 1.1em; line-height: 1.8; color: #333;">

**Reading a File**
- `file = open("data.txt", "r")`
- `content = file.read()`
- `file.close()`


**Better Way - Using With**
- `with open("data.txt", "r") as file:`
- `    content = file.read()`
- (file closes automatically)


**Writing to a File**
- `with open("output.txt", "w") as file:`
- `    file.write("Hello World!")`


**Appending to a File**
- `with open("log.txt", "a") as file:`
- `    file.write("New entry\n")`


**File Modes**
- `"r"` read mode
- `"w"` write mode (overwrites)
- `"a"` append mode (adds to end)
