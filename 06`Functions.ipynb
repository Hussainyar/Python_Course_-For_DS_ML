<!-- Slide 11: Function with Calculations -->
<section class="slide p-16 col" style="background-color: #0f172a; color: #f1f5f9;">
    <h2 class="mb-4" style="color: #38bdf8;">Step 7: Practical Example - Temperature Converter</h2>
    <pre><span class="keyword">def</span> celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * <span class="number">9</span>/<span class="number">5</span>) + <span class="number">32</span>
    <span class="keyword">return</span> fahrenheit

temp_c = <span class="number">25</span>
temp_f = celsius_to_fahrenheit(temp_c)
print(<span class="string">f"{temp_c}°C = {temp_f}°F"</span>)</pre>
    <p style="color: #86efac; font-family: 'Courier New', monospace; margin-top: 1rem;">

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

temp_c = 25
temp_f = celsius_to_fahrenheit(temp_c)
print(f"{temp_c}°C = {temp_f}°F")
```

<!-- Slide 12: List Operations in Functions -->
<section class="slide p-16 col" style="background-color: #0f172a; color: #f1f5f9;">
    <h2 class="mb-4" style="color: #38bdf8;">Step 8: Functions with Lists</h2>
    <pre><span class="keyword">def</span> calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    <span class="keyword">return</span> total / count

grades = [<span class="number">85</span>, <span class="number">90</span>, <span class="number">88</span>, <span class="number">92</span>]
avg = calculate_average(grades)
print(<span class="string">f"Average: {avg}"</span>)</pre>
    <p style="color: #86efac; font-family: 'Courier New', monospace; margin-top: 1rem;">
        Output:<br>
        <span style="color: #86efac;">Average: 88.75</span>

```python
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

grades = [85, 90, 88, 92]
avg = calculate_average(grades)
print(f"Average: {avg}")
```

<!-- Slide 15: Multiple Return Values -->
<section class="slide p-16 col" style="background-color: #0f172a; color: #f1f5f9;">
    <h2 class="mb-4" style="color: #38bdf8;">Step 10: Multiple Return Values (Tuples)</h2>
    <pre><span class="keyword">def</span> get_student_stats(grade_list):
    highest = max(grade_list)
    lowest = min(grade_list)
    <span class="keyword">return</span> highest, lowest

best, worst = get_student_stats([<span class="number">85</span>, <span class="number">92</span>, <span class="number">78</span>, <span class="number">95</span>])
print(<span class="string">f"Best: {best}, Worst: {worst}"</span>)</pre>
    <p style="color: #86efac; font-family: 'Courier New', monospace; margin-top: 1rem;">
        Output:<br>
        <span style="color: #86efac;">Best: 95, Worst: 78</span>
    </p>
</section>


```python
def get_student_stats(grade_list):
    highest = max(grade_list)
    lowest = min(grade_list)
    return highest, lowest

best, worst = get_student_stats([85, 92, 78, 95])
print(f"Best: {best}, Worst: {worst}")
```
