For Loops - The Basics

<h4 style="color: #38bdf8; margin-top: 0;">Syntax:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">for variable in sequence:
    # Code to repeat
    print(variable)</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">Explanation:</h4>
<ul style="margin-left: 20px;">
  <li><strong>for:</strong> Keyword that starts the loop</li>
  <li><strong>variable:</strong> Takes each value from the sequence</li>
  <li><strong>in:</strong> Connects variable to sequence</li>
  <li><strong>sequence:</strong> What we loop through (list, range, string)</li>
  <li><strong>Colon (:)</strong> Marks the start of the code block</li>
  <li><strong>Indentation:</strong> The code inside must be indented</li>
</ul>

<h4 style="color: #38bdf8; margin-top: 20px;">Example:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">for fruit in ['apple', 'banana', 'orange']:
    print(f'I like {fruit}')

# Output:
# I like apple
# I like banana
# I like orange</pre>

````python
# Basic For Loop Example


# Loop through a list of fruits

fruits = ['apple', 'banana', 'orange', 'mango', 'grape']

for fruit in fruits:
    print(fruit)
````

````python
word = "Python"

for letter in word:
    print(letter)
````

````python
even_numbers = []

for i in range(1, 11):
    if i % 2 == 0:
        even_numbers.append(i)

print(even_numbers)
````

````python
numbers = [1, 2, 3, 4, 5]
doubled = []

for num in numbers:
    doubled.append(num * 2)

print(doubled)
````

````python
numbers = [2, 8, 1, 6, 9, 3]
count = 0

for num in numbers:
    if num > 5:
        count += 1

print("Count:", count)
````

````python
numbers = [4, 7, 1, 9, 2]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
````

<h4 style="color: #38bdf8; margin-top: 0;">What is range()?</h4>
<p>range() generates a sequence of numbers. Perfect when you know how many times to loop.</p>

<h5 style="color: #64748b;">1Ô∏è‚É£ range(stop) - Count from 0 to (stop-1)</h5>

    print(i)  # Prints: 2, 3, 4, 5</pre>

While Loops
</div>

<div style="padding: 40px; background: white; color: #334155; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; line-height: 1.8; font-size: 1.05em;">

<h4 style="color: #38bdf8; margin-top: 0;">What is a While Loop?</h4>
<ul style="margin-left: 20px;">
  <li>Repeats code <strong>as long as</strong> a condition is TRUE</li>
  <li>Different from for loops: you don't know how many times to repeat</li>
  <li>Stops when the condition becomes FALSE</li>
</ul>

<h4 style="color: #38bdf8; margin-top: 20px;">Syntax:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">while condition:
    # Code to repeat
    # Update the condition somehow</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">Example:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">count = 0
while count < 5:
    print(f'Count: {count}')
    count = count + 1  # Update count
    
# Output: Count: 0, 1, 2, 3, 4</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">‚ö†Ô∏è Important: Avoid Infinite Loops!</h4>
<p style="color: #dc2626;"><strong>Always update the condition variable!</strong> Otherwise the loop never stops.</p>

</div>

````python
total = 0
i = 1

while i <= 5:
    total = total + i
    i += 1

print("Total:", total)
````

````python
even_numbers = []
i = 1

while i <= 10:
    if i % 2 == 0:
        even_numbers.append(i)
    i += 1

print(even_numbers)
````

````python
numbers = [1, 2, 3, 4, 5]
doubled = []
i = 0

while i < len(numbers):
    doubled.append(numbers[i] * 2)
    i += 1

print(doubled)
````

Loop Control: break & continue
</div>

<div style="padding: 40px; background: white; color: #334155; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; line-height: 1.8; font-size: 1.05em;">

<h4 style="color: #38bdf8; margin-top: 0;">break - Stop the Loop</h4>
<p>Exits the loop immediately when something happens:</p>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">for i in range(10):
    if i == 5:
        break  # Exit loop when i equals 5
    print(i)
# Output: 0, 1, 2, 3, 4</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">continue - Skip to Next Iteration</h4>
<p>Skips the rest of the current iteration and goes to the next:</p>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">for i in range(5):
    if i == 2:
        continue  # Skip when i equals 2
    print(i)
# Output: 0, 1, 3, 4</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">Real-World Use Cases:</h4>
<ul style="margin-left: 20px;">
  <li><strong>break:</strong> Stop searching when item is found</li>
  <li><strong>continue:</strong> Skip invalid data or unwanted records</li>
</ul>

````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

````python
numbers = [1, 3, 5, 7, 9, 11]

for num in numbers:
    if num > 7:
        break
    print(num)
````

````python
for i in range(3):
    for j in range(5):
        if j == 2:
            break
        print(f"i={i}, j={j}")
````

````python
# Find a number in a list
numbers = [2, 4, 6, 8, 10, 12, 14]
target = 10

for num in numbers:
    print(f"Checking: {num}")
    if num == target:
        print(f"Found {target}! Exiting loop.")
        break
else:
    print("Number not found!")
````

````python
# Skip even numbers
for i in range(1, 11):
    if i % 2 == 0:  # If i is even
        continue  # Skip to next iteration
    print(f"Odd number: {i}")
````

## üéØ Array Indexing & Slicing

**Indexing (similar to Python lists):**
- `arr[0]` - First element
- `arr[-1]` - Last element
- `arr[2]` - Third element

**Slicing:**
- `arr[1:4]` - Elements from index 1 to 3
- `arr[:3]` - First 3 elements
- `arr[3:]` - From index 3 to end
- `arr[::2]` - Every second element

> ‚ö†Ô∏è **Important:** Slicing creates a view, not a copy! Changes affect the original array.

**Creating Special 2D Arrays:**

- `np.ones((2, 3))` - 2x3 array of ones

## ‚ûï Array Arithmetic Operations

**Element-wise Operations (Vectorized!):**

| Operation | Description |
|-----------|-------------|
| `arr + 5` | Add 5 to all elements |
| `arr - 3` | Subtract 3 from all elements |
| `arr * 2` | Multiply all elements by 2 |
| `arr / 4` | Divide all elements by 4 |
| `arr ** 2` | Square all elements |

**Array-to-Array Operations:**
- `arr1 + arr2` (element-wise addition)

> üöÄ **This is MUCH faster than looping in Python!**


## üßÆ Mathematical Functions

**Universal Functions (ufuncs):**

| Function | Description |
|----------|-------------|
| `np.sqrt(arr)` | Square root of each element |
| `np.exp(arr)` | Exponential (e^x) |
| `np.log(arr)` | Natural logarithm |
| `np.sin(arr)` | Sine function |
| `np.cos(arr)` | Cosine function |
| `np.abs(arr)` | Absolute value |
| `np.round(arr, 2)` | Round to 2 decimal places |



## üìà Statistical Operations

**Common Statistical Functions:**

| Function | Description |
|----------|-------------|
| `np.mean(arr)` | Average of all elements |
| `np.median(arr)` | Median value |
| `np.std(arr)` | Standard deviation |
| `np.var(arr)` | Variance |
| `np.min(arr)` | Minimum value |
| `np.max(arr)` | Maximum value |
| `np.sum(arr)` | Sum of all elements |
| `np.argmin(arr)` | Index of minimum value |
| `np.argmax(arr)` | Index of maximum value |

> üí° **Tip:** Use `axis=0` or `axis=1` for operations on rows or columns in 2D arrays!

Broadcasting allows NumPy to perform operations on arrays of different shapes automatically!

**Example:**
```python
arr = np.array([1, 2, 3])
arr + 10  # Adds 10 to each element!
# Result: [11, 12, 13]
```

**Rules:**
- NumPy compares shapes element-wise
- Two dimensions are compatible when they are equal or one of them is 1
- Makes code cleaner and faster!

## ‚úÖ Boolean Indexing & Filtering

**Conditional Selection:**
```python
arr[arr > 5]  # Select elements greater than 5
```

**Creating Boolean Masks:**
- `arr > 5` - Returns True/False array
- `arr == 10` - Elements equal to 10
- `arr != 0` - Elements not equal to 0

**Combining Conditions:**
- `(arr > 5) & (arr < 10)` - AND condition
- `(arr < 3) | (arr > 7)` - OR condition

> ‚ö° **This is extremely powerful for data filtering in real projects!**

arr.reshape(3, 4)  # Reshape to 3 rows, 4 columns

**Flatten:** Convert multi-dimensional array to 1D
```python
arr.flatten()
```

**Transpose:** Swap rows and columns
```python
arr.T  # or arr.transpose()
```

**Auto-calculate dimension:**
```python
arr.reshape(2, -1)  # -1 means "figure it out"
