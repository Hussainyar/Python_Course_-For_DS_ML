For Loops - The Basics
</div>

<div style="padding: 40px; background: white; color: #334155; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; line-height: 1.8; font-size: 1.05em;">

<h4 style="color: #38bdf8; margin-top: 0;">Syntax:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">for variable in sequence:
    # Code to repeat
    print(variable)</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">Explanation:</h4>
<ul style="margin-left: 20px;">
  <li><strong>for:</strong> Keyword that starts the loop</li>
  <li><strong>variable:</strong> Takes each value from the sequence</li>
  <li><strong>in:</strong> Connects variable to sequence</li>
  <li><strong>sequence:</strong> What we loop through (list, range, string)</li>
  <li><strong>Colon (:)</strong> Marks the start of the code block</li>
  <li><strong>Indentation:</strong> The code inside must be indented</li>
</ul>

<h4 style="color: #38bdf8; margin-top: 20px;">Example:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">for fruit in ['apple', 'banana', 'orange']:
    print(f'I like {fruit}')

# Output:
# I like apple
# I like banana
# I like orange</pre>

```python
# Basic For Loop Example
print("=" * 50)
print("BASIC FOR LOOP EXAMPLE")
print("=" * 50)

# Loop through numbers 1 to 5
for i in range(1, 6):
    print(f"Number: {i}")

print("\n" + "=" * 50)
print("LOOP THROUGH A LIST")
print("=" * 50)

# Loop through a list of fruits
fruits = ['apple', 'banana', 'orange', 'mango', 'grape']
for fruit in fruits:
    print(f"I like {fruit}")
```

<h4 style="color: #38bdf8; margin-top: 0;">What is range()?</h4>
<p>range() generates a sequence of numbers. Perfect when you know how many times to loop.</p>

<h4 style="color: #38bdf8; margin-top: 20px;">Three ways to use range():</h4>

<h5 style="color: #64748b;">1Ô∏è‚É£ range(stop) - Count from 0 to (stop-1)</h5>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">for i in range(5):
    print(i)  # Prints: 0, 1, 2, 3, 4</pre>

<h5 style="color: #64748b; margin-top: 15px;">2Ô∏è‚É£ range(start, stop) - Count from start to (stop-1)</h5>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">for i in range(2, 6):
    print(i)  # Prints: 2, 3, 4, 5</pre>

<h5 style="color: #64748b; margin-top: 15px;">3Ô∏è‚É£ range(start, stop, step) - Count with steps</h5>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">for i in range(0, 10, 2):
    print(i)  # Prints: 0, 2, 4, 6, 8</pre>

</div>

<div style="background: #f8fafc; padding: 20px 40px; text-align: right; color: #94a3b8; font-size: 0.95em; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif;">

```python
# FOR LOOP WITH range()
print("\n" + "=" * 50)
print("range(5) - From 0 to 4")
print("=" * 50)
for i in range(5):
    print(i)

print("\n" + "=" * 50)
print("range(2, 8) - From 2 to 7")
print("=" * 50)
for i in range(2, 8):
    print(i)

print("\n" + "=" * 50)
print("range(0, 10, 2) - From 0 to 9, step by 2")
print("=" * 50)
for i in range(0, 10, 2):
    print(i)

print("\n" + "=" * 50)
print("range(10, 0, -1) - Countdown from 10 to 1")
print("=" * 50)
for i in range(10, 0, -1):
    print(i)
```

While Loops
</div>

<div style="padding: 40px; background: white; color: #334155; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; line-height: 1.8; font-size: 1.05em;">

<h4 style="color: #38bdf8; margin-top: 0;">What is a While Loop?</h4>
<ul style="margin-left: 20px;">
  <li>Repeats code <strong>as long as</strong> a condition is TRUE</li>
  <li>Different from for loops: you don't know how many times to repeat</li>
  <li>Stops when the condition becomes FALSE</li>
</ul>

<h4 style="color: #38bdf8; margin-top: 20px;">Syntax:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #334155;">while condition:
    # Code to repeat
    # Update the condition somehow</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">Example:</h4>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">count = 0
while count < 5:
    print(f'Count: {count}')
    count = count + 1  # Update count
    
# Output: Count: 0, 1, 2, 3, 4</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">‚ö†Ô∏è Important: Avoid Infinite Loops!</h4>
<p style="color: #dc2626;"><strong>Always update the condition variable!</strong> Otherwise the loop never stops.</p>

</div>

<div style="background: #f8fafc; padding: 20px 40px; text-align: right; color: #94a3b8; font-size: 0.95em; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif;">

```python
# WHILE LOOP EXAMPLE
print("\n" + "=" * 50)
print("WHILE LOOP EXAMPLE")
print("=" * 50)

# Count from 1 to 5
count = 1
while count <= 5:
    print(f"Count: {count}")
    count = count + 1

print("\n" + "=" * 50)
print("USER INPUT GAME")
print("=" * 50)

# Simple guessing game
secret_number = 7
guess = 0
attempts = 0

while guess != secret_number:
    guess = int(input("Guess a number between 1-10: "))
    attempts = attempts + 1

    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        print(f"Correct! You took {attempts} attempts.")
```

Loop Control: break & continue
</div>

<div style="padding: 40px; background: white; color: #334155; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; line-height: 1.8; font-size: 1.05em;">

<h4 style="color: #38bdf8; margin-top: 0;">break - Stop the Loop</h4>
<p>Exits the loop immediately when something happens:</p>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">for i in range(10):
    if i == 5:
        break  # Exit loop when i equals 5
    print(i)
# Output: 0, 1, 2, 3, 4</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">continue - Skip to Next Iteration</h4>
<p>Skips the rest of the current iteration and goes to the next:</p>
<pre style="background: #f8fafc; padding: 15px; border-radius: 4px; font-family: 'Courier New', monospace; color: #16a34a;">for i in range(5):
    if i == 2:
        continue  # Skip when i equals 2
    print(i)
# Output: 0, 1, 3, 4</pre>

<h4 style="color: #38bdf8; margin-top: 20px;">Real-World Use Cases:</h4>
<ul style="margin-left: 20px;">
  <li><strong>break:</strong> Stop searching when item is found</li>
  <li><strong>continue:</strong> Skip invalid data or unwanted records</li>
</ul>

</div>

<div style="background: #f8fafc; padding: 20px 40px; text-align: right; color: #94a3b8; font-size: 0.95em; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif;">

```python
# BREAK AND CONTINUE
print("\n" + "=" * 50)
print("BREAK STATEMENT - Exit Loop Early")
print("=" * 50)

# Find a number in a list
numbers = [2, 4, 6, 8, 10, 12, 14]
target = 10

for num in numbers:
    print(f"Checking: {num}")
    if num == target:
        print(f"Found {target}! Exiting loop.")
        break
else:
    print("Number not found!")

print("\n" + "=" * 50)
print("CONTINUE STATEMENT - Skip to Next Iteration")
print("=" * 50)

# Skip even numbers
for i in range(1, 11):
    if i % 2 == 0:  # If i is even
        continue  # Skip to next iteration
    print(f"Odd number: {i}")
```

```python
# LOOP THROUGH LISTS
print("\n" + "=" * 50)
print("SIMPLE LIST LOOP")
print("=" * 50)

students = ['Ahmed', 'Sara', 'Ali', 'Fatima', 'Hassan']
for student in students:
    print(f"Student: {student}")

print("\n" + "=" * 50)
print("ENUMERATE - Get Index and Value")
print("=" * 50)

for index, student in enumerate(students):
    print(f"{index + 1}. {student}")

print("\n" + "=" * 50)
print("MODIFY LIST WHILE LOOPING")
print("=" * 50)

scores = [85, 90, 78, 92, 88]
print(f"Original scores: {scores}")

# Add 5 points to each score
bonus_scores = []
for score in scores:
    bonus_scores.append(score + 5)

print(f"Scores with bonus: {bonus_scores}")
```
