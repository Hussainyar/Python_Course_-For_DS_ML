Variables in Python
</div>

<div style="padding: 40px; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; font-size: 1.1em; line-height: 1.8; color: #333;">

**What is a Variable?**
- A container that stores information
- Like a labeled box holding data
- You can create, read, and change variables


**Creating Variables**
- Use a name and assign a value with =
- Example: `name = "Ahmed"` or `age = 25`
- Names should be meaningful and simple


**Variable Naming Rules**
- Start with a letter or underscore (_)
- Use letters, numbers, and underscores only
- No spaces or special characters
- Good names: `my_name`, `student_age`, `total_score`


**Common Data Types**
- Integers: `age = 25` (whole numbers)
- Strings: `name = "Ali"` (text, use quotes)
- Floats: `height = 1.75` (decimal numbers)
- Booleans: `is_active = True` (True or False)
